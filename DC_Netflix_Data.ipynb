## **Loading the data set into a dictionary.**

As evidence of this, they have provided us with the following information. For the years from 2011 to 2020, the average movie durations are 103, 101, 99, 100, 100, 95, 95, 96, 93, and 90, respectively.

```python
# Create the years and durations lists
# import datetime

# year = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
year = [ i for i in range(2011,2021)]


durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]

movie_dict = {'year': year, 'durations': durations}



# movie_dict = {2011: 103, 2012: 101, 2013: 99, 2014: 100, 2015: 100, 2016: 95, 2017: 95, 2018: 96, 2019: 93, 2020: 90}


# Print the dictionary
print(movie_dict)
```

Create DF

```python
# Import pandas under its usual alias
import pandas as pd

# Create a DataFrame from the dictionary
durations_df = pd.DataFrame(movie_dict)

# Print the DataFrame
print(durations_df)
```

```python
# Import matplotlib.pyplot under its usual alias and create a figure
import matplotlib.pyplot as plt
fig = plt.figure()

# Draw a line plot of release_years and durations
plt.plot(durations_df['year'], durations_df['durations'])

# Create a title
plt.title("Netflix Movie Durations 2011-2020")

# Show the plot
plt.show()
```

###**Filtering out the whole CSV converted DataFrame**

Rows with "Movie" and the columns - "title", "country", "genre", "release_year", "duration"


```python
# Subset the DataFrame for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df["type"] == "Movie"]

# Select only the columns of interest
netflix_movies_col_subset = netflix_df[["title", "country", "genre", "release_year", "duration"]]

# Print the first five rows of the new DataFrame
print(netflix_movies_col_subset.iloc[0:5])
```

### Scatter Plot of year vs. duration


```python
import matplotlib.pyplot as plt

# Create a figure and increase the figure size
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus year
plt.scatter(netflix_df["release_year"], netflix_df["duration"])

# Create a title
plt.title("Movie Duration by Year of Release")

# Show the plot
plt.show()
```

Filtering the whole data set to "Movies" and "Short-Movies"

```python
# Filter for durations shorter than 60 minutes
short_movies = netflix_movies_col_subset[netflix_movies_col_subset["duration"] < 60]

# netflix_df_movies_only = netflix_df[netflix_df["type"] == "Movie"]

# Print the first 20 rows of short_movies
print(short_movies.iloc[0:20])
```

### Applying colours to different genres
Iterate over rows of netflix_movies_col_subset

```python
# Define an empty list
colors = []
# Iterate over rows of netflix_movies_col_subset
for x, row in netflix_movies_col_subset.iterrows() : 
  if row['genre'] == 'Children' :
    colors.append('red')
  elif row['genre'] == 'Documentaries' :
    colors.append('blue')
  elif row['genre'] == 'Stand-Up' :
    colors.append('green') 
  else:  
    colors.append('black')
# Inspect the first 10 values in your list
print(colors[0:11])
```

**Scatter Plot with colour scale**

```python
import matplotlib.pyplot as plt
# Set the figure style and initalize a new figure
fig = plt.figure(figsize=(12,8))
# Create a scatter plot of duration versus release_year
plt.scatter(netflix_movies_col_subset.release_year, netflix_movies_col_subset.duration, c =colors)
# Create a title and axis labels
plt.title("Movie duration by year of release") 
plt.xlabel("Release year") 
plt.ylabel("Duration (min)")
```
